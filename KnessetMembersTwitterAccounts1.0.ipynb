```python
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
import time
import pandas as pd
```

```python
driver = webdriver.Chrome()
url = 'http://main.knesset.gov.il/mk/Pages/current.aspx?pg=mklist'
driver.get(url)
```

```python
info = []


def get_twitter(twitter):
    try:  
      #  twitter = driver.find_element_by_xpath("""//*[@id="ctl00_PlaceHolderMain_dcpDesktop_ctl00_IconTwitter"]""").get_attribute('href')
         twitter = driver.find_element_by_id('ctl00_PlaceHolderMain_dcpDesktop_ctl00_IconTwitter').get_attribute('href')

    except NoSuchElementException:
         twitter = 'none'
    return(twitter)


def get_info():
    mail = driver.find_element_by_id('ctl00_PlaceHolderMain_dcpDesktop_ctl00_EmailDiv').text.replace('דואר אלקטרוני:\n', '').lower()     
    phone = driver.find_elements_by_class_name('MKIdInfo')[0].text.replace('טלפון בכנסת:\n', '')
    name = driver.find_element_by_class_name('BreadcrumbPageTitle').text.replace('חבר הכנסת ','').replace('חברת הכנסת ','')
    
    twitter = 'none'
    twitter = get_twitter(twitter)
    
    member = {}
    member['name'] = name
    member['phone'] = phone
    member['mail'] = mail
    member['twitter'] = twitter
    info.append(member)
```

```python
start_time = time.time()


Num_of_members = len (driver.find_elements_by_class_name("MKLobbyMKNameDiv")) 


for index in range(Num_of_members) :
    driver.find_elements_by_class_name("MKLobbyMKNameDiv")[index].click()
    driver.implicitly_wait(5)
                 
    get_info()
    
    driver.back()
    driver.implicitly_wait(5)
    #if index == 10:
     #   break 


print("run time --- %s seconds ---" % (time.time() - start_time))
```

```python
df = pd.DataFrame(info)
df = df[['name','mail','phone','twitter']]

df
```
